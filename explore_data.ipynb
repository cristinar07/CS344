```python
fig, axs = plt.subplots(10, 2, figsize=(14,40))

for i in range(len(numerical_variables)):
    
    sns.kdeplot(data=data, x=numerical_variables[i], hue='Cover_Type', palette='turbo', fill=True, ax=axs[i][0])
    sns.boxplot(data=data, y=numerical_variables[i], x='Cover_Type', palette='turbo', fliersize=2, ax=axs[i][1])
    
    axs[i][0].set_xlabel(numerical_variables_labels[i], fontsize=14)
    axs[i][1].set_ylabel(numerical_variables_labels[i], fontsize=14)
    axs[i][1].set_xlabel('Cover Type', fontsize=14)
    axs[i][0].set_ylabel('Density', fontsize=14)

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.3)

plt.show()
```

```python
# one hot encoded data
one_hot_encoded_data = pd.get_dummies(data, columns = ['Cover_Type'])
```

```python
# determine correlation

corr = one_hot_encoded_data.corr()
plt.figure(figsize=(20,20)) 
sns.heatmap(corr, cmap='RdYlGn_r')
plt.show()
```
